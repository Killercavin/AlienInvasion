```python
from matplotlib import pyplot as plt

plt.title('SQUARE VALUES', fontsize=15, color='blue')
plt.ylabel('Square of values', fontsize=12.5)
plt.xlabel('Values', fontsize=12.5)
# plt.style.use('classic')

x_values = range(1, 1000)
y_values = [x**2 for x in x_values]
plt.plot(x_values, y_values, color='maroon')
plt.savefig('squares_plot.png', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 1000)
y_values = [x**3 for x in x_values]

plt.title('CUBIC VALUES', fontsize=15, color='blue')
plt.ylabel('Cube of values', fontsize=12.5)
plt.xlabel('Values', fontsize=12.5)

plt.plot(x_values, y_values)
plt.savefig('cube_plot.png', bbox_inches='tight')
plt.show()
```
